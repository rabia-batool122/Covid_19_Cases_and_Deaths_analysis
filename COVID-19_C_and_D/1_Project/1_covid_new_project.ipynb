# What are the total number of deaths in each Continent by Countries?

- Clean-up total_deaths column
- Clean-up total_cases column
- Calculate number of deaths by Continent
- Calculate number cases by Country


```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("bolkonsky/covid19")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


path = '/Users/macbook/Covid_19_C_and_D_analysis/1_COVID-19_C_and_D/owid-covid-data.csv'
df = pd.read_csv(path)
df.copy()

df['date'] = pd.to_datetime(df['date'])
df_continents = df.continent.unique().tolist()
df_continents = df.continent.dropna().to_frame()
df['total_deaths'] = df['total_deaths'].fillna(0).astype(int)
df['total_cases'] = df['total_cases'].fillna(0).astype(int)
```

```python
df_continent_country = df.groupby(['continent', 'location'])[['total_cases', 'total_deaths']].sum()
df_continent_country = df_continent_country.reset_index()
df_continent_country
```

```python
df_continent_country_cases = df_continent_country.sort_values(by='total_cases', ascending=False).drop(columns='total_deaths').head(10)
df_continent_country_cases
```

```python
df_continent_country_deaths = df_continent_country.sort_values(by='total_deaths', ascending=False).drop(columns='total_cases').head(10)
df_continent_country_deaths
```

```python
df_continent = df.groupby('continent')[['total_cases', 'total_deaths']].sum().reset_index()
df_continent
```
